```python
import numpy as np
import scipy.sparse
import scipy.sparse.linalg
import matplotlib.pyplot as plt
```

```python
# Load the data
alphas_77 = np.load(r'C:\Users\lg330\Downloads\hs_tomography\hs_tomography\alphas_77.npy')
y_77 = np.load(r'C:\Users\lg330\Downloads\hs_tomography\hs_tomography\y_77.npy')
alphas_195 = np.load(r'C:\Users\lg330\Downloads\hs_tomography\hs_tomography\alphas_195.npy')
y_195 = np.load(r'C:\Users\lg330\Downloads\hs_tomography\hs_tomography\y_195.npy')
```

```python
# Load the precomputed weight matrix for the smaller version
X_77 = np.load(r'C:\Users\lg330\Downloads\hs_tomography\hs_tomography\X_77.npy', allow_pickle=True)
X_77_sparse = scipy.sparse.csc_matrix(X_77.all())
```

```python
# Find the number of non-zero entries in X_77
num_nonzero_entries = X_77_sparse.nnz
sparsity = num_nonzero_entries / (X_77_sparse.shape[0] * X_77_sparse.shape[1])
print(f'Number of non-zero entries in X_77: {num_nonzero_entries}')
print(f'Sparsity of X_77: {sparsity}')
```

```python
# Solve the linear equation using lsqr
atol = 1e-05
btol = 1e-05
result = scipy.sparse.linalg.lsqr(X_77_sparse, y_77, atol=atol, btol=btol)
beta_77 = result[0]
```

```python
# Reshape beta to form the tomogram image
M = 77
tomogram_77 = beta_77.reshape((M, M))
```

```python
# Plot the reconstructed tomogram
plt.imshow(tomogram_77, cmap='gray')
plt.title('Reconstructed Tomogram (77x77)')
plt.colorbar()
plt.show()
```
